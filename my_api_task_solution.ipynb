### Introduction
Data used in this analysis was derived from GitHub via JSON API resquest. The primary source of the data is Openfootball. The request made was for the __2019/2020 English Premier League__ season matches dataset. 

The intend for this exercise was to access the dataset and determine by analysis information which are not immediately visible when viewing league standings table. Consequently, the aim is primarily not to reproduce the league standings table but instead, to show among other things miniscule and or granular details possible, such as follows: 

* the team that sufferred the worsre defeat and the team that inflicted it
* the team with the highest numbers of goals scored
* the highest number of goals scored
* the teams that had more goals scored away than at home
* the team that may have won the league
* the team with the highest number of draws 
* the overall goals for and against for each team during the season
* the team with the highest number losses at home and away from home
* the team with the highest number of wins at home and away
* number of losses
* number of draws
* number of wins
* the total numbers of goals scored during the season
* the average number of goals scored by each team
* the total numbers of goals scored by each team
* conceeded goals home and away
* the team that scored more goals away than at home
* top five contenders (UEFA & Europa Bound teams)
* last three contenders (relagation bound teams)

### Transforming the request into a dataframe

```python
match_date=[]
team_1=[]
team_2=[]
ft_score1=[]
ft_score2=[]
for m in data["matches"]:
    match_date.append(m["date"])
    team_1.append(m["team1"])
    team_2.append(m["team2"])
    ft_score1.append(m["score"]["ft"][0])
    ft_score2.append(m["score"]["ft"][1])
    
```
